```python
import csv
from astropy.io import fits
import numpy as np
import sys
import math
import glob
```

```python
def combining_script(file,output_name):
    
    '''This script takes an input fits file of fluxes from different filters and combines them into magnitudes
    all together. It will average those that have more than one measurement.'''
    
    hdul = fits.open(file)

    RA = []
    Dec = []
    RA2000 = []
    Dec2000 = []
    avezmag = []
    aveymag = []
    avejmag = []
    avehmag = []
    avekmag = []
    avezobsmag = []
    aveyobsmag = []
    avejobsmag = []
    avehobsmag = []
    avekobsmag = []
    avezaper = []
    aveyaper = []
    avejaper = []
    avehaper = []
    avekaper = []
    avezapererr = []
    aveyapererr = []
    avejapererr = []
    avehapererr = []
    avekapererr = []
    
    output_fits = output_name + '.fits'
    output_reg = output_name + '.reg'

    for x in range(len(hdul[1].data)):
        """Getting the RA and Dec"""
        if hdul[1].data[x]['RA_T2'] != '':
            RA.append(hdul[1].data[x]['RA_T2'])
            Dec.append(hdul[1].data[x]['Dec_T2'])
            RA2000.append(hdul[1].data[x]['RA2000_T2'])
            Dec2000.append(hdul[1].data[x]['Dec2000_T2'])
        elif hdul[1].data[x]['RA_T3'] != '':
            RA.append(hdul[1].data[x]['RA_T3'])
            Dec.append(hdul[1].data[x]['Dec_T3'])
            RA2000.append(hdul[1].data[x]['RA2000_T3'])
            Dec2000.append(hdul[1].data[x]['Dec2000_T3'])
        elif hdul[1].data[x]['RA_T6'] != '':
            RA.append(hdul[1].data[x]['RA_T6'])
            Dec.append(hdul[1].data[x]['Dec_T6'])
            RA2000.append(hdul[1].data[x]['RA2000_T6'])
            Dec2000.append(hdul[1].data[x]['Dec2000_T6'])
        elif hdul[1].data[x]['RA_T7'] != '':
            RA.append(hdul[1].data[x]['RA_T7'])
            Dec.append(hdul[1].data[x]['Dec_T7'])
            RA2000.append(hdul[1].data[x]['RA2000_T7'])
            Dec2000.append(hdul[1].data[x]['Dec2000_T7'])

        """Getting the Magnitudes"""
        '''Start with Correct Mag'''
        t2zmag = hdul[1].data[x]['Mag_Correct_Z_T2']
        t3zmag = hdul[1].data[x]['Offset_Mag_Z_T3']
        t6zmag = hdul[1].data[x]['Mag_Correct_Z_T6']
        t7zmag = hdul[1].data[x]['Mag_Correct_Z_T7']
        t2zflux = 10**(0.4*t2zmag)
        t3zflux = 10**(0.4*t3zmag)
        t6zflux = 10**(0.4*t6zmag)
        t7zflux = 10**(0.4*t7zmag)
        avezflux = np.nanmean((t2zflux,t3zflux,t6zflux,t7zflux))
        avezmag.append(2.5*math.log(avezflux,10))
        t3ymag = hdul[1].data[x]['Mag_Correct_Y_T3']
        t6ymag = hdul[1].data[x]['Mag_Correct_Y_T6']
        t7ymag = hdul[1].data[x]['Mag_Correct_Y_T7']
        t3yflux = 10**(0.4*t3ymag)
        t6yflux = 10**(0.4*t6ymag)
        t7yflux = 10**(0.4*t7ymag)
        aveyflux = np.nanmean((t3yflux,t6yflux,t7yflux))
        aveymag.append(2.5*math.log(aveyflux,10))
        t2jmag = hdul[1].data[x]['Mag_Correct_J_T2']
        t3jmag = hdul[1].data[x]['Mag_Correct_J_T3']
        t6jmag = hdul[1].data[x]['Mag_Correct_J_T6']
        t7jmag = hdul[1].data[x]['Mag_Correct_J_T7']
        t2jflux = 10**(0.4*t2jmag)
        t3jflux = 10**(0.4*t3jmag)
        t6jflux = 10**(0.4*t6jmag)
        t7jflux = 10**(0.4*t7jmag)
        avejflux = np.nanmean((t2jflux,t3jflux,t6jflux,t7jflux))
        avejmag.append(2.5*math.log(avejflux,10))
        t2hmag = hdul[1].data[x]['Mag_Correct_H_T2']
        t3hmag = hdul[1].data[x]['Mag_Correct_H_T3']
        t6hmag = hdul[1].data[x]['Mag_Correct_H_T6']
        t7hmag = hdul[1].data[x]['Mag_Correct_H_T7']
        t2hflux = 10**(0.4*t2hmag)
        t3hflux = 10**(0.4*t3hmag)
        t6hflux = 10**(0.4*t6hmag)
        t7hflux = 10**(0.4*t7hmag)
        avehflux = np.nanmean((t2hflux,t3hflux,t6hflux,t7hflux))
        avehmag.append(2.5*math.log(avehflux,10))
        t2kmag = hdul[1].data[x]['Mag_Correct_K_T2']
        t3kmag = hdul[1].data[x]['Mag_Correct_K_T3']
        t6kmag = hdul[1].data[x]['Mag_Correct_K_T6']
        t7kmag = hdul[1].data[x]['Mag_Correct_K_T7']
        t2kflux = 10**(0.4*t2kmag)
        t3kflux = 10**(0.4*t3kmag)
        t6kflux = 10**(0.4*t6kmag)
        t7kflux = 10**(0.4*t7kmag)
        avekflux = np.nanmean((t2kflux,t3kflux,t6kflux,t7kflux))
        avekmag.append(2.5*math.log(avekflux,10))

        '''Do Observed Mag'''
        t2zobsmag = hdul[1].data[x]['Mag_Obs_Z_T2']
        t3zobsmag = hdul[1].data[x]['Mag_Obs_Z_T3']
        t6zobsmag = hdul[1].data[x]['Mag_Obs_Z_T6']
        t7zobsmag = hdul[1].data[x]['Mag_Obs_Z_T7']
        t2zobsflux = 10**(0.4*t2zobsmag)
        t3zobsflux = 10**(0.4*t3zobsmag)
        t6zobsflux = 10**(0.4*t6zobsmag)
        t7zobsflux = 10**(0.4*t7zobsmag)
        avezobsflux = np.nanmean((t2zobsflux,t3zobsflux,t6zobsflux,t7zobsflux))
        avezobsmag.append(2.5*math.log(avezobsflux,10))
        t3yobsmag = hdul[1].data[x]['Mag_Obs_Y_T3']
        t6yobsmag = hdul[1].data[x]['Mag_Obs_Y_T6']
        t7yobsmag = hdul[1].data[x]['Mag_Obs_Y_T7']
        t3yobsflux = 10**(0.4*t3yobsmag)
        t6yobsflux = 10**(0.4*t6yobsmag)
        t7yobsflux = 10**(0.4*t7yobsmag)
        aveyobsflux = np.nanmean((t3yobsflux,t6yobsflux,t7yobsflux))
        aveyobsmag.append(2.5*math.log(aveyobsflux,10))
        t2jobsmag = hdul[1].data[x]['Mag_Obs_J_T2']
        t3jobsmag = hdul[1].data[x]['Mag_Obs_J_T3']
        t6jobsmag = hdul[1].data[x]['Mag_Obs_J_T6']
        t7jobsmag = hdul[1].data[x]['Mag_Obs_J_T7']
        t2jobsflux = 10**(0.4*t2jobsmag)
        t3jobsflux = 10**(0.4*t3jobsmag)
        t6jobsflux = 10**(0.4*t6jobsmag)
        t7jobsflux = 10**(0.4*t7jobsmag)
        avejobsflux = np.nanmean((t2jobsflux,t3jobsflux,t6jobsflux,t7jobsflux))
        avejobsmag.append(2.5*math.log(avejobsflux,10))
        t2hobsmag = hdul[1].data[x]['Mag_Obs_H_T2']
        t3hobsmag = hdul[1].data[x]['Mag_Obs_H_T3']
        t6hobsmag = hdul[1].data[x]['Mag_Obs_H_T6']
        t7hobsmag = hdul[1].data[x]['Mag_Obs_H_T7']
        t2hobsflux = 10**(0.4*t2hobsmag)
        t3hobsflux = 10**(0.4*t3hobsmag)
        t6hobsflux = 10**(0.4*t6hobsmag)
        t7hobsflux = 10**(0.4*t7hobsmag)
        avehobsflux = np.nanmean((t2hobsflux,t3hobsflux,t6hobsflux,t7hobsflux))
        avehobsmag.append(2.5*math.log(avehobsflux,10))
        t2kobsmag = hdul[1].data[x]['Mag_Obs_K_T2']
        t3kobsmag = hdul[1].data[x]['Mag_Obs_K_T3']
        t6kobsmag = hdul[1].data[x]['Mag_Obs_K_T6']
        t7kobsmag = hdul[1].data[x]['Mag_Obs_K_T7']
        t2kobsflux = 10**(0.4*t2kobsmag)
        t3kobsflux = 10**(0.4*t3kobsmag)
        t6kobsflux = 10**(0.4*t6kobsmag)
        t7kobsflux = 10**(0.4*t7kobsmag)
        avekobsflux = np.nanmean((t2kobsflux,t3kobsflux,t6kobsflux,t7kobsflux))
        avekobsmag.append(2.5*math.log(avekobsflux,10))

        '''Do Aper flux, and then error'''
        t2zaper = hdul[1].data[x]['Aper_flux_Z_T2']
        t3zaper = hdul[1].data[x]['Aper_flux_Z_T3']
        t6zaper = hdul[1].data[x]['Aper_flux_Z_T6']
        t7zaper = hdul[1].data[x]['Aper_flux_Z_T7']
        avezaper.append(np.nanmean((t2zaper,t3zaper,t6zaper,t7zaper)))
        t3yaper = hdul[1].data[x]['Aper_flux_Y_T3']
        t6yaper = hdul[1].data[x]['Aper_flux_Y_T6']
        t7yaper = hdul[1].data[x]['Aper_flux_Y_T7']
        aveyaper.append(np.nanmean((t3yaper,t6yaper,t7yaper)))
        t2japer = hdul[1].data[x]['Aper_flux_J_T2']
        t3japer = hdul[1].data[x]['Aper_flux_J_T3']
        t6japer = hdul[1].data[x]['Aper_flux_J_T6']
        t7japer = hdul[1].data[x]['Aper_flux_J_T7']
        avejaper.append(np.nanmean((t2japer,t3japer,t6japer,t7japer)))
        t2haper = hdul[1].data[x]['Aper_flux_H_T2']
        t3haper = hdul[1].data[x]['Aper_flux_H_T3']
        t6haper = hdul[1].data[x]['Aper_flux_H_T6']
        t7haper = hdul[1].data[x]['Aper_flux_H_T7']
        avehaper.append(np.nanmean((t2haper,t3haper,t6haper,t7haper)))
        t2kaper = hdul[1].data[x]['Aper_flux_K_T2']
        t3kaper = hdul[1].data[x]['Aper_flux_K_T3']
        t6kaper = hdul[1].data[x]['Aper_flux_K_T6']
        t7kaper = hdul[1].data[x]['Aper_flux_K_T7']
        avekaper.append(np.nanmean((t2kaper,t3kaper,t6kaper,t7kaper)))

        t2zapererr = hdul[1].data[x]['Aper_flux_Z_err_T2']
        t3zapererr = hdul[1].data[x]['Aper_flux_Z_err_T3']
        t6zapererr = hdul[1].data[x]['Aper_flux_Z_err_T6']
        t7zapererr = hdul[1].data[x]['Aper_flux_Z_err_T7']
        avezapererr.append(np.nanmean((t2zapererr,t3zapererr,t6zapererr,t7zapererr)))
        t3yapererr = hdul[1].data[x]['Aper_flux_Y_err_T3']
        t6yapererr = hdul[1].data[x]['Aper_flux_Y_err_T6']
        t7yapererr = hdul[1].data[x]['Aper_flux_Y_err_T7']
        aveyapererr.append(np.nanmean((t3yapererr,t6yapererr,t7yapererr)))
        t2japererr = hdul[1].data[x]['Aper_flux_J_err_T2']
        t3japererr = hdul[1].data[x]['Aper_flux_J_err_T3']
        t6japererr = hdul[1].data[x]['Aper_flux_J_err_T6']
        t7japererr = hdul[1].data[x]['Aper_flux_J_err_T7']
        avejapererr.append(np.nanmean((t2japererr,t3japererr,t6japererr,t7japererr)))
        t2hapererr = hdul[1].data[x]['Aper_flux_H_err_T2']
        t3hapererr = hdul[1].data[x]['Aper_flux_H_err_T3']
        t6hapererr = hdul[1].data[x]['Aper_flux_H_err_T6']
        t7hapererr = hdul[1].data[x]['Aper_flux_H_err_T7']
        avehapererr.append(np.nanmean((t2hapererr,t3hapererr,t6hapererr,t7hapererr)))
        t2kapererr = hdul[1].data[x]['Aper_flux_K_err_T2']
        t3kapererr = hdul[1].data[x]['Aper_flux_K_err_T3']
        t6kapererr = hdul[1].data[x]['Aper_flux_K_err_T6']
        t7kapererr = hdul[1].data[x]['Aper_flux_K_err_T7']
        avekapererr.append(np.nanmean((t2kapererr,t3kapererr,t6kapererr,t7kapererr)))
        
    col1 = fits.Column(name='RA',format='11A',unit='dms',array=RA)
    col2 = fits.Column(name='Dec',format='11A',unit='dms',array=Dec)
    col3 = fits.Column(name='RA2000',format='D',unit='radians',array=RA2000)
    col4 = fits.Column(name='Dec2000',format='D',unit='radians',array=Dec2000)
    col5 = fits.Column(name='Mag_Z',format='D',unit='mag',array=avezmag)
    col6 = fits.Column(name='Obs_Mag_Z',format='D',unit='mag',array=avezobsmag)
    col7 = fits.Column(name='Aper_Z',format='E',unit='ADU',array=avezaper)
    col8 = fits.Column(name='Aper_Z_err',format='E',unit='ADU',array=avezapererr)
    col9 = fits.Column(name='Mag_Y',format='D',unit='mag',array=aveymag)
    col10 = fits.Column(name='Obs_Mag_Y',format='D',unit='mag',array=aveyobsmag)
    col11 = fits.Column(name='Aper_Y',format='E',unit='ADU',array=aveyaper)
    col12 = fits.Column(name='Aper_Y_err',format='E',unit='ADU',array=aveyapererr)
    col13 = fits.Column(name='Mag_J',format='D',unit='mag',array=avejmag)
    col14 = fits.Column(name='Obs_Mag_J',format='D',unit='mag',array=avejobsmag)
    col15 = fits.Column(name='Aper_J',format='E',unit='ADU',array=avejaper)
    col16 = fits.Column(name='Aper_J_err',format='E',unit='ADU',array=avejapererr)
    col17 = fits.Column(name='Mag_H',format='D',unit='mag',array=avehmag)
    col18 = fits.Column(name='Obs_Mag_H',format='D',unit='mag',array=avehobsmag)
    col19 = fits.Column(name='Aper_H',format='E',unit='ADU',array=avehaper)
    col20 = fits.Column(name='Aper_H_err',format='E',unit='ADU',array=avehapererr)
    col21 = fits.Column(name='Mag_K',format='D',unit='mag',array=avekmag)
    col22 = fits.Column(name='Obs_Mag_K',format='D',unit='mag',array=avekobsmag)
    col23 = fits.Column(name='Aper_K',format='E',unit='ADU',array=avekaper)
    col24 = fits.Column(name='Aper_K_err',format='E',unit='ADU',array=avekapererr)
    
    cols = fits.ColDefs([col1,col2,col3,col4,col5,col6,col7,col8,col9,col10,col11,col12,col13,col14,col15,col16,col17,col18,col19,col20,col21,col22,col23,col24])
    
    '''Here we make the fits file'''

    data = fits.BinTableHDU.from_columns(cols)
    data.writeto(output_fits)

    '''Now we make the reg file'''

    pos_data = []

    for v in range(len(RA)):

        pos_data.append([RA[v],Dec[v]])

    with open(output_reg, 'w', newline='') as csvfile:
        reg_file = csv.writer(csvfile, delimiter=',', quotechar=' ', quoting=csv.QUOTE_ALL)
        for w in range(len(pos_data)):
            reg_file.writerow([pos_data[w][0] , pos_data[w][1]])
```

```python
combining_script('combined_3filters.fits','combined_output_3filt_redo')
```
